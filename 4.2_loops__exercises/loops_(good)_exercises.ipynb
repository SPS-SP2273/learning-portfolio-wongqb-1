```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

for i in range(len(co2_data)-1):
    year = []
    CO2_level = []
    year.append(co2_data[i][0])
    CO2_level.append(co2_data[i][1])
    year.append(co2_data[i+1][0])
    CO2_level.append(co2_data[i+1][1])
    change = CO2_level[1] - CO2_level[0]
    if np.abs(change) >= 3:
        print(f'{year[0]} to {year[1]}: Change: {change:.4f}')
```
