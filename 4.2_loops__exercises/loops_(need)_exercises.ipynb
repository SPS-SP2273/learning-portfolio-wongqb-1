```python
import numpy as np
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
temperatures_celciusArr = np.array(temperatures_celsius)
temperatures_fahrenheitArr = 1.8 * temperatures_celciusArr + 32

for i in range(temperatures_celciusArr.size):
    print(f'Celcius: {temperatures_celciusArr[i]} \t Fahrenheit: {temperatures_fahrenheitArr[i]}')
```

```python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
```

```python
for num1 in [1,2,3,4,5]:
    print(f'{num1} : ', end='\t')

    for num2 in [1,2,3,4,5]:
        prod = num1 * num2
        print(f'{prod}', end='\t')
    
    print('\n')
```
