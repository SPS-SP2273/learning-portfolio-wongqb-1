```python
def convert_celcius(temperature_celcius, target_scale='Fahrenheit'):
    if target_scale == 'Fahrenheit':
        return temperature_celcius * 1.8 + 32
    elif target_scale == 'Kelvin':
        return temperature_celcius + 273.15
    else:
        return 'Invalid input.'

convert_celcius(102, 'Fahrenheit')
```

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale='Celcius'):
    if target_scale == 'Celcius':
        return (temperature_fahrenheit - 32) / 1.8
    elif target_scale == 'Kelvin':
        return (temperature_fahrenheit - 32) / 1.8 + 273.15
    else:
        return 'Invalid input.'

convert_fahrenheit(215.6, 'Celcius')
```

```python
def convert_temperature(temperature, source_scale, target_scale):
    if source_scale == 'Celcius':
        return convert_celcius(temperature, target_scale)
    elif source_scale == 'Fahrenheit':
        return convert_fahrenheit(temperature, target_scale)
    else:
        return 'Invalid input.'

convert_temperature(215.6, 'Fahrenheit', 'Celcius')
```
