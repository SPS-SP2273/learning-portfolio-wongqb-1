```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'

greeting("Batman")
```

The code above works as intended since:
1. If `name != 'Batman'` then the `if` code block will not be executed and `return f'Hello {name}!'` will be executed.
2. If `name == 'Batman'` then the `if` code block will be executed, which will reach the `return` statement and terminate the flow of the definition of the function, hence `return f'Hello {name}'` is not run. 

```python
def bmi(weight, height):
    BMI = weight / (height ** 2)
    if BMI < 18.5:
        return 'Underweight'
    elif BMI < 24.9:
        return 'Normal weight'
    elif BMI < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
    
bmi(60, 1.74)
```

```python
def factorial(n):
    if n < 0:
        return 'This program does not handle negative numbers.'
    result = 1
    while n != 0:
        result *= n
        n -= 1
    return result

factorial(48)
```
