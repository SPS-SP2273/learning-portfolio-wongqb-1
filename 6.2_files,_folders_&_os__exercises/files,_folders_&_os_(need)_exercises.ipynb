```python
import os
import glob
import shutil

def rename_my_file(pathStr):
    
    monthDict = {
        'jan': '01', 'feb': '02', 'mar': '03',
        'apr': '04', 'may': '05', 'jun': '06',
        'jul': '07', 'aug': '08', 'sep': '09',
        'oct': '10', 'nov': '11', 'dec': '12'
    }  

    path = os.path.split(pathStr)   # Returns parent directory and file (with extension)
    splittedPath = path[1].split('-')
    finalPathStr = os.path.join(path[0], f'{monthDict[splittedPath[0]]}-{splittedPath[1]}') # Using an f-string and the os.path.join() function to make sure directory separators are correct
    return finalPathStr

fileList = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
newFileList = []

for i in range(len(fileList)):
    newFileList.append(
        rename_my_file(fileList[i])
    )
    print(f'{fileList[i]:<20} ----> {newFileList[i]}')
```

```python
# We first copy the file as new name, then we delete the original file.
# However the above is not necessary since the os class provides a rename() function.

filesArr = glob.glob(r'os-collaboration-exercise-data\*.txt')
for i in range(len(filesArr)):
    try:
        os.rename(filesArr[i], 
              f'{rename_my_file(filesArr[i])}')
        print(f'Done {filesArr[i]} ----> {rename_my_file(filesArr[i])}')
    except:
        print('An unexpected error has occurred. Maybe the files are already renamed?')

```
