|#|Term|Description|
|-|----|-----------|
|1|Function|An object that does something, for example printing to the shell, or add two numbers, etc.|
|2|Arguments|The parameters required / optionally needed for the function to run according to one's needs|
|3|Comments|Text placed in code to explain the code, and of which is ignored by the interpreter|
|4|Indentations|The whitespaces that are used to describe structures such as if-else blocks for the interpreter|
|5|Packages|Additional installable items that provide non-default functions, objects, etc.|
|6|`.`|Used to access the member of a class, for example arr.size is an attribute which is the size of the array arr.|
|7|`:`|Used after if-else statements, for loops, etc. to mark the beginning of the block|
|8|`[]`|Specifies a list|
|9|`[[],[]]`|Specifies a list of lists|
|10|`{}`|Specifies a dictionary|
