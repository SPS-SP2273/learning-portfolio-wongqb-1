```python
import numpy as np
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

x_1Array = np.array(x_1)
x_2Array = np.array(x_2)
x_1Mean = x_1Array.mean()
x_2Mean = x_2Array.mean()
x_1Std = x_1Array.std()
x_2Std = x_2Array.std()

print('Mean for x_1:', x_1Mean)
print('Mean for x_2:', x_2Mean)
print('Standard deviation for x_1:', x_1Std)
print('Standard deviation for x_2:', x_2Std)
print('Pearson correlation coefficient')
```
