Recall that the Pearson correlation coefficient is defined as $$\begin{align*}r &= \dfrac{\sum_{i=1}^n (x_i - \mu_X)(y_i -\mu_Y)}{\sqrt{\sum_{i=1}^{n}(x_i - \mu_X)^2\sum_{i=1}^{n}(y_i - \mu_Y)^2}}
\end{align*}$$

```python
import numpy as np

x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

x_1Array = np.array(x_1)
x_2Array = np.array(x_2)
x_1Mean = x_1Array.mean()
x_2Mean = x_2Array.mean()
x_1Std = x_1Array.std()
x_2Std = x_2Array.std()
sumMeanProduct = np.float32(0)
sumMeanSquare1 = np.float32(0)
sumMeanSquare2 = np.float32(0)

for i in range(x_1Array.size):
    try:
        sumMeanProduct += (x_1Array[i] - x_1Mean) * (x_2Array[i] - x_2Mean)
        sumMeanSquare1 += (x_1Array[i] - x_1Mean) ** 2
        sumMeanSquare2 += (x_2Array[i] - x_2Mean) ** 2
    except:
        print('Arrays not the same size!')

pearson = sumMeanProduct / np.sqrt(sumMeanSquare1 * sumMeanSquare2)

print(f'Mean for x_1: {x_1Mean}')
print(f'Mean for x_2: {x_2Mean}')
print(f'Standard deviation for x_1: {x_1Std:.4f}')
print(f'Standard deviation for x_2: {x_2Std:.4f}')
print(f'Pearson correlation coefficient: {pearson}')
```
